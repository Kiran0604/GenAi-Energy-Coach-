```python
# Load and process all PDFs
!pip install chromadb
from langchain_community.vectorstores import Chroma
all_docs = []
for filename in os.listdir(pdf_folder_path):
    if filename.endswith(".pdf"):
        file_path = os.path.join(pdf_folder_path, filename)
        loader = PyMuPDFLoader(file_path)
        documents = loader.load()
        splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
        chunks = splitter.split_documents(documents)
        all_docs.extend(chunks)

print(f"✅ Total chunks: {len(all_docs)}")

# Create and persist vectorstore
vectorstore = Chroma.from_documents(
    documents=all_docs,
    embedding=embedding,
    persist_directory=persist_directory
)
vectorstore.persist()
print("✅ Vector DB created and saved locally.")
```

```python
# Reload vectorstore and create retriever
vectorstore = Chroma(
    persist_directory=persist_directory,
    embedding_function=embedding
)
retriever = vectorstore.as_retriever()
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
# Define prompt template

from langchain.prompts import PromptTemplate
prompt_template = PromptTemplate(
    input_variables=["question", "context"],
    template=(
        "You are an AI-powered virtual assistant helping industries in India understand energy efficiency schemes, government incentives, and policies.\n"
        "Provide accurate, practical, and concise guidance based strictly on the provided context.\n"
        "Use simple, professional language suitable for engineers, plant operators, or MSME managers.\n\n"
        "Question: {question}\n\n"
        "Context: {context}\n\n"
        "Answer:"
    )
)
# Initialize your LLM here (replace with your model)
# Example: llm = HuggingFaceHub(repo_id="some-llm-repo")
# For demonstration, 'llm' should be defined before this point
qa_chain = LLMChain(llm=llm, prompt=prompt_template)

# Function to answer questions using RAG
def answer_question_with_rag(question):
    retrieved_docs = retriever.get_relevant_documents(question)
    context = " ".join([doc.page_content for doc in retrieved_docs])
    response = qa_chain.run(question=question, context=context)
    return response
```

```python
# Python: Simulate data acquisition (if no hardware)
import pandas as pd
import numpy as np
import time
from datetime import datetime

def simulate_sensor_data(n=1000):
    timestamps = pd.date_range(datetime.now(), periods=n, freq='min')
    current = np.random.normal(5, 1, n)  # Amps
    voltage = np.random.normal(230, 5, n)  # Volts
    power = current * voltage  # Watts
    energy = np.cumsum(power) / 60  # Wh (approx)
    df = pd.DataFrame({'timestamp': timestamps, 'current': current, 'voltage': voltage, 'power': power, 'energy': energy})
    df.to_csv('energy_data.csv', index=False)
    return df

# Simulate and save data
sensor_df = simulate_sensor_data(1440)  # 1 day, 1-min interval
sensor_df.head()
```

## 2. Data Logging & Preprocessing

- Load real or simulated data
- Prepare for ML/GenAI analysis

```python
# Load data
import pandas as pd
sensor_df = pd.read_csv('energy_data.csv', parse_dates=['timestamp'])
sensor_df.head()
```

## 3. ML Models: Forecasting & Anomaly Detection

- LSTM/ARIMA for forecasting
- IsolationForest for anomaly detection

```python
# Forecasting with ARIMA (simple example)
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

power_series = sensor_df['power']
model = ARIMA(power_series, order=(2,1,2))
model_fit = model.fit()
forecast = model_fit.forecast(steps=60)

plt.figure(figsize=(10,4))
plt.plot(power_series[-200:], label='Actual')
plt.plot(range(len(power_series), len(power_series)+60), forecast, label='Forecast')
plt.legend()
plt.title('Power Forecast (ARIMA)')
plt.show()
```

```python
# Anomaly Detection
from sklearn.ensemble import IsolationForest
sensor_df['anomaly'] = IsolationForest(contamination=0.01).fit_predict(sensor_df[['power', 'current', 'voltage']])

plt.figure(figsize=(10,4))
plt.plot(sensor_df['timestamp'], sensor_df['power'], label='Power')
plt.scatter(sensor_df['timestamp'][sensor_df['anomaly']==-1], sensor_df['power'][sensor_df['anomaly']==-1], color='red', label='Anomaly')
plt.legend()
plt.title('Anomaly Detection')
plt.show()
```

## 4. GenAI Integration: Natural Language Insights & Q&A

- Use LangChain/OpenAI/HuggingFace for GenAI
- Generate reports, answer queries, ROI suggestions

```python
# Install and import GenAI libraries (LangChain, OpenAI, etc.)
!pip install langchain openai
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Example: Natural language report generation
llm = OpenAI(temperature=0.2)
prompt = PromptTemplate(
    input_variables=["summary"],
    template="""
    You are an energy efficiency coach. Given this summary of machine data:
    {summary}
    Generate a concise, actionable report for plant managers.
    """
)
llm_chain = LLMChain(llm=llm, prompt=prompt)

summary = f"Average power: {sensor_df['power'].mean():.2f} W, Max: {sensor_df['power'].max():.2f} W, Anomalies: {sum(sensor_df['anomaly']==-1)}."
report = llm_chain.run(summary=summary)
print(report)
```

```python
# Q&A Example
user_query = "Suggest ways to reduce boiler energy consumption."
context = summary  # In practice, use more context
prompt = PromptTemplate(
    input_variables=["question", "context"],
    template="""
    You are an AI coach. Question: {question}\nContext: {context}\nAnswer in 3 bullet points.
    """
)
llm_chain = LLMChain(llm=llm, prompt=prompt)
answer = llm_chain.run(question=user_query, context=context)
print(answer)
```

## 5. IEM & Operational Optimization

- Calculate KPIs: Energy/Product Ratio, Idle Time Waste
- Suggest off-peak schedules, investment ROI

```python
# Example KPI calculation
sensor_df['shift'] = ((sensor_df['timestamp'].dt.hour // 8) + 1)
energy_per_shift = sensor_df.groupby('shift')['energy'].max().diff().fillna(sensor_df.groupby('shift')['energy'].max())
print('Energy per shift (Wh):', energy_per_shift.values)

# Example: ROI calculation
old_motor_power = 1000  # W
new_motor_power = 800   # W
hours_per_year = 3000
cost_per_kWh = 8  # INR
savings = (old_motor_power - new_motor_power) * hours_per_year / 1000 * cost_per_kWh
print(f"Estimated annual savings by replacing motor: ₹{savings:,.0f}")
```
